```python
import sys
import numpy as np
import scipy.sparse as sparse
from scipy.sparse.linalg import spsolve
import random

from sklearn.preprocessing import MinMaxScaler

import implicit
```

```python
sparse_item_user = sparse.csr_matrix((dataset['rating_x'].astype(float),(dataset['placeID'], dataset['userID_x'])))
```

```python
sparse_user_item = sparse.csr_matrix((dataset['rating_x'].astype(float),(dataset['userID_x'], dataset['placeID'])))
```

```python
model = implicit.als.AlternatingLeastSquares(factors=20,regularization=0.1,iterations=200)
```

```python
alpha_val =15
data_conf = (sparse_item_user*alpha_val).astype('double')
```

```python
model.fit(data_conf)
```

```python
item_id = 135085
n_similar = 5
similar = model.similar_items(item_id,n_similar)
```

```python
for item in similar:
    idx,score = item
    print(dataset.name.loc[dataset.placeID == idx].iloc[0])
```

## Find user Recommendation

```python
user_id = 134
recommeded = model.recommend(user_id, sparse_user_item)
```

```python
restaurant = []
scores = []
```

```python
for item in recommeded:
    idx,score = item
    restaurant.append(dataset.name.loc[dataset.placeID==idx].iloc[0])
    scores.append(score)
```

```python
print(pd.DataFrame({"restaurant":restaurant, "scores":scores}))
```
